```python
from pprint import pprint
from Bio import Entrez
from parse_pubmed import PubMedArticleSet PMCArticleSet
import pickle
# import sqlite3
```

```python
Entrez.email = 'nick.morley@hotmail.com'

query = 'amyloid cascade'
```

```python
# Pubmed abstracts

with Entrez.esearch(db='pubmed', term=query, retmax=2) as handle:
    record = Entrez.read(handle)
    pmids = record['IdList']
    
print(len(pmids))
```

```python
%%time

handle = Entrez.efetch(db='pubmed', id=pmids, rettype='full', retmode='xml')
xml = handle.read()

with open('data/amyloid_cascade_pmc.xml', 'w', encoding='utf-8') as f:
    f.write(xml)
```

```python
print(xml)
```

```python
# Pubmed Central

with Entrez.esearch(db='pmc', term=query, retmax=1500) as handle:
    record = Entrez.read(handle)
    pmcids = record['IdList']
    
print(len(pmcids))
```

```python
%%time

handle = Entrez.efetch(db='pmc', id=pmcids, rettype='full', retmode='xml')
xml = handle.read()  # PMC article set

with open('data/nootropic_pmc.xml', 'w', encoding='utf-8') as f:
    f.write(xml)
```

```python
with open('data/nootropic_pmc.xml', 'r', encoding='utf-8') as f:
    xml = f.read()
```
